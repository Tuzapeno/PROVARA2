# Questão 1 (vale 6 pontos)
Considerando as funções listadas a seguir faça: **(a)** o gráfico das três funções; **(b)** o passo a passo algébrico, em latex, para encontrar o ângulo de inclinação da reta tangente no ponto $x=5$ para cada das funções:

1. $
f(x) = (x^2 - 4)(x^3 - 8x-1)
$
2. $
f(x) = \sin(x^3)\cos(x)
$
3.$
f(x) = x^2e^{-2x}\ln(x)
$

Se você usar o sympy, terá que fazer com que ele imprima todos os passos necessários para resolver cada uma das operações, regras, ou tabelas que aplicar para resolver o problema.

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio.


```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import math

x = sp.Symbol('x')

def f1(x):
  return ((x**2) - 4)*((x**3) - 8*x - 1)

def f2(x):
  return sp.sin(x**3)*sp.cos(x)

def f3(x):
  if x > 0:
    return (x**2)*sp.exp(-2*x)*sp.log(x)

# Valores de X
valoresX = [x for x in np.linspace(0, 10, 100)]

# Valores para a função
funcaoY1 = [f1(x) for x in valoresX]
funcaoY2 = [f2(x) for x in valoresX]
funcaoY3 = [f3(x) for x in valoresX]

# Funcao 1
plt.plot(valoresX, funcaoY1, color="green", linestyle="-", label="f1(x)")

# Funcao 2
plt.plot(valoresX, funcaoY2, color="purple", linestyle="--", label="f2(x)")

# Funcao 3
plt.plot(valoresX, funcaoY3, color="blue", linestyle=":", label="f3(x)")

# Funções simbólicas
f1_sym = (x**2 - 4) * (x**3 - 8 * x - 1)
f2_sym = sp.sin(x**3) * sp.cos(x)
f3_sym = (x**2) * sp.exp(-2 * x) * sp.log(x)

# Derivadas
f1_der = sp.diff(f1_sym, x)
f2_der = sp.diff(f2_sym, x)
f3_der = sp.diff(f3_sym, x)

# Derivadas no ponto x=5
f1_der_5 = f1_der.subs(x, 5)
f2_der_5 = f2_der.subs(x, 5)
f3_der_5 = f3_der.subs(x, 5)

# Ângulo de inclinação em graus
angle1 = math.degrees(math.atan(f1_der_5))
angle2 = math.degrees(math.atan(f2_der_5))
angle3 = math.degrees(math.atan(f3_der_5))

# Imprimir os resultados
print("f1(x) =", f1_sym)
print("f1'(x) =", f1_der)
print("f1'(5) =", f1_der_5)
print("ângulo de inclinação =", angle1, "graus")
print("")

print("f2(x) =", f2_sym)
print("f2'(x) =", f2_der)
print("f2'(5) =", f2_der_5.evalf())
print("ângulo de inclinação =", angle2, "graus")
print("")

print("f3(x) =", f3_sym)
print("f3'(x) =", f3_der)
print("f3'(5) =", f3_der_5.evalf())
print("ângulo de inclinação =", angle3, "graus")


# Legendas
plt.legend()
```

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Como sempre, você pode usar qualquer ferramenta disponível na internet para lhe auxiliar a resolver este problema. Lembre-se, contudo, que o sympy é uma boa alternativa. 

Em todas as questões você pode usar apenas os métodos, regras, técnicas e algoritmos que utilizamos na nossa disciplina, em sala, ou no material de apoio. Lembrem-se estamos limitados as bibliotecas sympy, numpy e matplotlib.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
import numpy as np
import sympy as sp

x = sp.Symbol('x')

# Funcao
f = (7 * sp.sin(x)) - (2 * sp.cos(x))

# Derivada da funcao
derivada = sp.diff(f, x)

pontosCriticos = sp.solve(f, x)
minimo = []
for value in pontosCriticos:
  if value > -10 and value < 10:
    minimo.append(value)
    break
  
# Encontrar o primeiro ponto de mínimo
min_value = f.subs(x, minimo[0]).evalf()

print("Ponto mínimo no intervalo -10 <= x <= 10: x =", round(minimo[0].evalf(), 4), ", f(x) =", round(min_value.evalf(), 4))
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+2y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import numpy as np
import sympy as sp
import plotly.graph_objs as go

# Definindo a função
def f(x, y):
    return np.sin(x * y) + np.cos(x + 2 * y)

# Criando uma matriz de valores da função no intervalo especificado
valoresX = np.linspace(0, 5, 100)
valoresY = np.linspace(0, 5, 100)
X, Y = np.meshgrid(valoresX, valoresY)
Z = f(X, Y)

# Encontrando os índices dos pontos de máximo e mínimo
max_index = np.unravel_index(np.argmax(Z), Z.shape)
min_index = np.unravel_index(np.argmin(Z), Z.shape)

# Coordenadas dos pontos de máximo e mínimo
maximo = (X[max_index], Y[max_index], Z[max_index])
minimo = (X[min_index], Y[min_index], Z[min_index])


figura = go.Figure()

figura.add_trace(go.Surface(x=X, y=Y, z=Z, colorscale='Viridis', opacity=0.8))

figura.add_trace(go.Scatter3d(x=[maximo[0]], y=[maximo[1]], z=[maximo[2]],
                           mode='markers', marker=dict(size=8, color='red'), name='Máximo'))

figura.add_trace(go.Scatter3d(x=[minimo[0]], y=[minimo[1]], z=[minimo[2]],
                           mode='markers', marker=dict(size=8, color='blue'), name='Mínimo'))

figura.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='f(x, y)'),
                  title='Gráfico de f(x, y) = sin(xy) + cos(x + 2y)')

figura.show()
```
